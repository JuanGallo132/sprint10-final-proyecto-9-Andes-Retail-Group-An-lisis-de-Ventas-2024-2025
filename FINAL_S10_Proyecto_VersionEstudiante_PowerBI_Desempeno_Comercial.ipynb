### Narrativa del Dashboard (Modelo SCQA)

#### 🖥️ Vista General (Overview)

**S (Situación):**  
Andes Retail Group cerró el bienio 2024–2025 operando en tres mercados de la región andina, con métricas de ingresos, margen y volumen como principales indicadores de desempeño.

**C (Complicación):**  
2025 registró una caída generalizada en las tres métricas clave respecto a 2024. Adicionalmente, el análisis temporal revela un valle recurrente entre junio y agosto, lo que indica que la baja estacional de invierno no es un evento aislado, sino un patrón estructural del negocio.

**Q (Pregunta):**   
¿El deterioro de 2025 responde únicamente a la estacionalidad, o existen segmentos y categorías que están amplificando la caída?

**A (Respuesta):**  
La caída es transversal, pero no homogénea. La segmentación por tipo de cliente y el análisis mensual permiten identificar dónde se concentra la pérdida y qué palancas activar para mitigarla en el corto plazo.

---

#### 🔎 Vista Detalle

**S (Situación):**  
El análisis detallado desagrega el rendimiento por categoría de producto, segmento de cliente, país y estación del año para los periodos 2024–2025.

**C (Complicación):**  
El tercer trimestre (julio–septiembre) es consistentemente el más débil en ingresos y margen. Perú concentra el 39% del ingreso total y supera a Colombia en un 60%, generando una dependencia geográfica significativa. Las categorías de Deportes y Electrónica presentan el mayor potencial de recuperación estacional, pero sin intervención planificada, replican el mismo patrón de caída año tras año.

**Q (Pregunta):**  
¿Cómo puede el equipo de Marketing y Ventas anticiparse al Qtr3 para reducir el impacto estacional y diversificar la dependencia de Perú?

**A (Respuesta):**  
Activar campañas focalizadas en Deportes y Electrónica durante julio–septiembre, y desarrollar estrategias de crecimiento en Colombia para equilibrar la concentración geográfica de ingresos.
